### (a) (4 points) Tuning curve for one neuron

Fit a cosine tuning curve for electrode 17. On the same plot, plot the average firing rate for electrode 17 to each of the 8 directions and the fitted tuning curve.  The average firing rate should be computed in a window from [250, 500] ms after trial onset. Reach angle should be on the x-axis and firing rate on the y-axis. 

Hint: You may reuse code from HW #2 about tuning curve fitting to do this question.  We have provided two functions: calculateFR() and tuning() which ought simplify this question.  Take a look at what they do before calling them.  The function tuning returns the tuning curve of a neuron (i.e., the parameters of the tuning curve $c_0$, $c_1$, $\theta_0$ such that $f(\theta)= c_0+c_1 cos(\theta-\theta_0)$

```python
def calculateFR(i, neuron_idx=16, window=[250,500]):
    # Calculates the firing rate of neuron_idx on trial i in the given window range.
    # Output units are spikes/s
    return float(len(R[i]['spikeRaster'][neuron_idx,window[0]:window[1]].todense().nonzero()[1])) / (window[1] - window[0]) * 1000

def tuning(thetas, meanFRs):
    # Calculates the parameters of a tuning curve.
    F = np.concatenate((np.ones((np.size(thetas), 1)),np.matrix(np.sin(thetas)).T,np.matrix(np.cos(thetas)).T),axis = 1)
    
    ks = np.linalg.pinv(F)*np.matrix(meanFRs).T
    c0 = ks[0]
    pd = np.arctan2(ks[1],ks[2])
    c1 = ks[1]/np.sin(pd)
    
    c0 = np.asarray(c0)[0][0]
    c1 = np.asarray(c1)[0][0]
    pd = np.asarray(pd)[0][0]
    return [c0,c1,pd]
```

```python
plt.figure(figsize=(7,5))

#====================================================#
# YOUR CODE HERE:
#   Fit the cosine tuning curve for electrode 17 and plot
#     1) the average firing rate for electrode 17 to each of the 8 directions and 
#     2) the fitted tuning curve.
#====================================================#
num_trials = R.shape[0]
num_directions = 8

targets = np.zeros((num_trials, 3, 1))
for trial in range(num_trials):
    targets[trial] = R[trial]['target']
unique_target_locations = np.unique(targets, axis=0)

# Remove center location because have 9 locations
for loc in range(unique_target_locations.shape[0]):
    #print(unique_target_locations[i])
    if np.array_equal(unique_target_locations[loc], np.array([[0],[0],[-70]])):
        unique_target_locations = np.delete(unique_target_locations, loc, 0)
        break
        
# Change target directions to thetas
theta_params = np.zeros(num_directions)
for direc in range(num_directions):
    theta_params[direc] = np.arctan2(unique_target_locations[direc,1], unique_target_locations[direc,0])
    if theta_params[direc] < 0:
        theta_params[direc] += np.pi*2

# Get mean firing rates
firing_rates = np.zeros(num_directions)
FR_counts = np.zeros(num_directions)
for trial in range(num_trials):
    index = 0
    for target in unique_target_locations:
        if np.array_equal(R[trial]['target'], target):
            firing_rates[index] += calculateFR(trial)
            FR_counts[index] += 1
        index += 1

mean_FRs = firing_rates / FR_counts

plt.plot(theta_params * 180.0 / np.pi, mean_FRs, 'o')


[c0,c1,theta0] = tuning(theta_params, mean_FRs)
theta = np.linspace(0, np.pi*2, 60)

plt.plot(theta*180.0/np.pi, c0+c1*np.cos(theta - theta0))
#====================================================#
# END YOUR CODE
#====================================================#   

plt.xlabel('Reach angle')
plt.ylabel('Firing rate')
plt.xlim([0,360])
plt.show()
```

For the decoder in part (e), what is the average mean-square error in position per trial? (For each time bin in a trial, calculate the squared Euclidean distance between the decoded position and the true hand position.
```python
#====================================================#
# YOUR CODE HERE:
#   Calculate the mean-squared error between the decoded
#   hand position and the true hand position.  Average
#   the squared errors across time; then average the squared
#   errors across trials.
#====================================================#
MSEs = []

for trial in range(400, num_trials):
    true_position = R[trial]['cursorPos'][0:2,0::25]
    decoded_position = positions_testdata[trial - 400]
    if true_position.shape[1] > decoded_position.shape[1]:
        diff = (true_position[:, 0:decoded_position.shape[1]]-decoded_position)**2
    elif true_position.shape[1] < decoded_position.shape[1]:
        diff = (true_position-decoded_position[:,0:true_position.shape[1]])**2
    elif true_position.shape[1] == decoded_position.shape[1]:
        diff = (true_position-decoded_position)**2
    mse_time = np.mean(diff[0] + diff[1])
    MSEs.append(mse_time)
    
mse = np.mean(MSEs)
print(mse)
#====================================================#
# END YOUR CODE
#====================================================#
```
