### (b) (7 points)  Fit a Wiener filter and decode activity.

Fit the Wiener filter using the data matrix you generated in part(a). Plot, as you did in question 2(g), the decoded trajectories for each test trial. (Hint: to decode the first 3 velocities of trial i at times 25 ms, 50 ms, and 75 ms, since 100 ms have not yet occurred in the trial, you will have to use the last bins of spiking activity in trial $i−1$.)


```python
#====================================================#
# YOUR CODE HERE:
#   Fit the Wiener filter and decode the trajectories for
#   each test trial.
#====================================================#
plt.figure(figsize=(7,7))

num_trials = R.shape[0]

#must remove first 3 columns
X_w_bin = X_bin[:,3:]

L_w = np.matmul(X_w_bin, scipy.linalg.pinv(Y_w_bin))

positions_testdata = []

#need to intialize previous Y for first test data index
Y = R[train_num-1]['spikeRaster']
Y_bin_prior = nsp.bin(Y, dt,'sum')

for trial in range(train_num, num_trials):
    Y = R[trial]['spikeRaster']
    Y_bin = nsp.bin(Y, dt,'sum')
    Y_bin_prior = Y_bin_prior[:,-3:] 
    Y_bin_hist = np.concatenate((Y_bin_prior, Y_bin), axis=1)
    X_bin_next = []
    for j in range(Y_bin.shape[1]):
        Y_bin_p = np.fliplr(Y_bin_hist[:,j:j+4]).T
        Y_bin_next = Y_bin_p.reshape(Y_bin_p.shape[0] * Y_bin_p.shape[1])
        Y_bin_next = np.hstack((Y_bin_next, 1))
        X_trial = np.dot(L_w, Y_bin_next)
        X_bin_next.append(X_trial)
    Y_bin_prior = Y_bin

    starting_point = np.array([[R[trial]['cursorPos'][0,0]], [R[trial]['cursorPos'][1,0]]])
    
    pos_intervals = np.concatenate((starting_point, (np.array(X_bin_next).T)*0.025), axis=1)
    positions = np.cumsum(pos_intervals, axis=1)
    positions_testdata.append(positions)
    
    plt.plot(positions[0], positions[1], 'o')
    
#====================================================#
# END YOUR CODE
#====================================================#   
```

### (c) (4 points) Calculate the mean-square error for the Wiener filter.

What is the average mean-square error in position per trial? Did the Wiener filter perform better or worse than the optimal linear estimator?

```python
#====================================================#
# YOUR CODE HERE:
#   Calculate the mean-squared error between the decoded
#   hand position and the true hand position.  Average
#   the squared errors across time; then average the squared
#   errors across trials.
#====================================================#
MSEs = []

for trial in range(400, num_trials):
    true_position = R[trial]['cursorPos'][0:2,0::25]
    decoded_position = positions_testdata[trial - 400]
    if true_position.shape[1] > decoded_position.shape[1]:
        diff = (true_position[:, 0:decoded_position.shape[1]]-decoded_position)**2
    elif true_position.shape[1] < decoded_position.shape[1]:
        diff = (true_position-decoded_position[:,0:true_position.shape[1]])**2
    elif true_position.shape[1] == decoded_position.shape[1]:
        diff = (true_position-decoded_position)**2
    mse_time = np.mean(diff[0] + diff[1])
    MSEs.append(mse_time)
    
mse = np.mean(MSEs)
print(mse)
#====================================================#
# END YOUR CODE
#====================================================#  
```
